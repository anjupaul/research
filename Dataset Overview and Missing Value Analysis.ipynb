Correlation analysis of both datasets.



```python
import pandas as pd
import yfinance as yf

# Load the dataset
ticker = 'AAPL'  # You can change this to any other ticker WOW.AX
data = yf.download(ticker, start='2000-01-01', end='2024-01-01')
data = data.dropna()

# 1. Basic Information
print("Basic Information:")
print(f"Number of rows: {data.shape[0]}")
print(f"Number of columns: {data.shape[1]}")
print("\nColumn Names and Data Types:")
print(data.dtypes)
print("\nNumber of Missing Values in Each Column:")
print(data.isnull().sum())

# 2. Descriptive Statistics
print("\nDescriptive Statistics:")
print(data.describe())

# 3. Data Distribution (Histograms)
import matplotlib.pyplot as plt

print("\nHistograms:")
data.hist(figsize=(10, 8))
plt.show()

# 4. Correlation Matrix
print("\nCorrelation Matrix:")
correlation_matrix = data.corr()
print(correlation_matrix)

# Optional: Plot the correlation matrix as a heatmap for better visualization
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()

# Optional: Save the summary to a text file
with open('dataset_summary.txt', 'w') as f:
    f.write("Basic Information:\n")
    f.write(f"Number of rows: {data.shape[0]}\n")
    f.write(f"Number of columns: {data.shape[1]}\n")
    f.write("\nColumn Names and Data Types:\n")
    f.write(f"{data.dtypes}\n")
    f.write("\nNumber of Missing Values in Each Column:\n")
    f.write(f"{data.isnull().sum()}\n")
    f.write("\nDescriptive Statistics:\n")
    f.write(f"{data.describe()}\n")
    f.write("\nCorrelation Matrix:\n")
    f.write(f"{correlation_matrix}\n")
```
