```python
import cv2
import os
import random

# Define a list of video file paths and corresponding output folder paths
video_paths = [
    '/content/drive/MyDrive/MLandAIAssignment2/Videos/Bike_artificial.MOV',
    '/content/drive/MyDrive/MLandAIAssignment2/Videos/Bike_cloudy.MOV',
    '/content/drive/MyDrive/MLandAIAssignment2/Videos/Bike_natural.MOV',
    '/content/drive/MyDrive/MLandAIAssignment2/Videos/Bike_shadow.MOV',
    '/content/drive/MyDrive/MLandAIAssignment2/Videos/Bike_sunny.MOV',
    '/content/drive/MyDrive/MLandAIAssignment2/Videos/Car_artificial.MOV',
    '/content/drive/MyDrive/MLandAIAssignment2/Videos/Car_cloudy.MOV',
    '/content/drive/MyDrive/MLandAIAssignment2/Videos/Car_natural.MOV',
    '/content/drive/MyDrive/MLandAIAssignment2/Videos/Car_shadow.MOV',
    '/content/drive/MyDrive/MLandAIAssignment2/Videos/Car_sunny.MOV'
]

output_folders = [
    '/content/drive/MyDrive/MLandAIAssignment2/Images/preprocessing1/Bike_artificial',
    '/content/drive/MyDrive/MLandAIAssignment2/Images/preprocessing1/Bike_cloudy',
    '/content/drive/MyDrive/MLandAIAssignment2/Images/preprocessing1/Bike_natural',
    '/content/drive/MyDrive/MLandAIAssignment2/Images/preprocessing1/Bike_shadow',
    '/content/drive/MyDrive/MLandAIAssignment2/Images/preprocessing1/Bike_sunny',
    '/content/drive/MyDrive/MLandAIAssignment2/Images/preprocessing1/Car_artificial',
    '/content/drive/MyDrive/MLandAIAssignment2/Images/preprocessing1/Car_cloudy',
    '/content/drive/MyDrive/MLandAIAssignment2/Images/preprocessing1/Car_natural',
    '/content/drive/MyDrive/MLandAIAssignment2/Images/preprocessing1/Car_shadow',
    '/content/drive/MyDrive/MLandAIAssignment2/Images/preprocessing1/Car_sunny'
]
num_frames_to_extract = 150

# Specify the desired width and height for resized frames
desired_width = 160
desired_height = 160

# Function to equalize the histogram of a color image
def equalize_hist_color(image_unchanged):
    eq_channels = []
    channels = cv2.split(image_unchanged)
    for ch in channels:
        eq_channels.append(cv2.equalizeHist(ch))

    eq_image = cv2.merge(eq_channels)
    return eq_image

# Function to apply Gaussian blur and weighted addition to an image
def enhance_image(image):
    # Apply Gaussian blur
    blurred_image = cv2.GaussianBlur(image, (0, 0), 3)

    # Perform weighted addition for enhancement
    enhanced_image = cv2.addWeighted(image, 1.5, blurred_image, -0.5, 0)

    return enhanced_image

for video_path, output_path in zip(video_paths, output_folders):
    # Create the output folder if it doesn't exist
    os.makedirs(output_path, exist_ok=True)

    # Open the video file
    vidcap = cv2.VideoCapture(video_path)
    success, image = vidcap.read()

    # Get the total number of frames in the video
    total_frames = int(vidcap.get(cv2.CAP_PROP_FRAME_COUNT))

    # Create a list of random frame indices to extract
    random_frame_indices = random.sample(range(total_frames), num_frames_to_extract)

    frame_count = 0

    while success:
        if frame_count in random_frame_indices:
            image_name = f'frame{frame_count}.PNG'
            image_path = os.path.join(output_path, image_name)
            equalized_frame = equalize_hist_color(image)
            enhanced_frame = enhance_image(equalized_frame)
            resized_frame = cv2.resize(enhanced_frame, (desired_width, desired_height))
            cv2.imwrite(image_path, resized_frame)
            print(f'Saved frame {frame_count} from {video_path} to {output_path}')

        success, image = vidcap.read()
        frame_count += 1

    # Release the video capture object for this video
    vidcap.release()

# Optionally, close any open windows
cv2.destroyAllWindows()
```
